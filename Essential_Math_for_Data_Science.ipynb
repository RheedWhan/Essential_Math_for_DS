```python
import matplotlib.pyplot as plt

def f(x):
    return 2 * x + 1

x = [2, 4, 5, 8]
y = [f(x_val) for x_val in x]  # Calculate f(x) for each value in x

plt.plot(x, y, marker='o', linestyle='-')  # Plot the points and connect with lines
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = 2*x + 1')
plt.grid(True)
plt.show()
```

**Charting linear functions in Python using SymPy**

```python
from sympy import *

x = symbols('x')
y = 2*x + 1

plot(y)
```

**Explanation**

1. `from sympy import *`: This line imports all functions and symbols from the SymPy library. SymPy is a Python library for symbolic mathematics.

2. `x = symbols('x')`: This line defines a symbol 'x' using the `symbols` function from SymPy. In mathematics, symbols are used to represent variables, and 'x' is a common symbol for a variable.

3. `y = 2*x + 1`: This line defines an algebraic expression 'y' using the symbol 'x'. In this case, it represents the function `f(x) = 2*x + 1`. This function takes the variable 'x', multiplies it by 2, and then adds 1.

4. `plot(y)`: This line uses the `plot` function from SymPy to create a plot of the expression 'y'. The function 'y' represents a straight line with a slope of 2 and a y-intercept of 1, which is the graph of the equation `y = 2*x + 1`.

So, in summary, this code uses SymPy to define a symbolic variable 'x' and an algebraic expression 'y' that represents the function `f(x) = 2*x + 1`. It then uses SymPy's plotting capabilities to generate a graphical representation of this function, which is a straight line on the Cartesian plane.

**Charting exponential functions in Python using SymPy**

```python
from sympy import *

x = symbols('x')
y = x**2 + 1

plot(y)
```

**Explanation**

1. `from sympy import *`: This line imports all functions and classes from the SymPy library. SymPy is a Python library for symbolic mathematics, which allows you to perform symbolic computations, manipulate algebraic expressions, and solve equations symbolically.

2. `x = symbols('x')`: This line defines a symbol 'x' using the `symbols` function from SymPy. In mathematics, symbols are used to represent variables, and here 'x' is treated as an algebraic variable.

3. `y = x**2 + 1`: This line defines an algebraic expression 'y' using the 'x' symbol. The expression is `x**2 + 1`, which represents a quadratic function. In mathematical notation, this is equivalent to \(y = x^2 + 1\).

4. `plot(y)`: This line plots the expression 'y' using SymPy's `plot` function. It takes the algebraic expression 'y' and generates a plot of this function. In this case, it will create a plot of the quadratic function \(y = x^2 + 1\).

So, in summary, the code defines a symbolic algebraic expression \(y = x^2 + 1\) and then plots this expression to visualize the graph of the quadratic function. The SymPy library is used for symbolic mathematics, making it easier to work with mathematical expressions and equations in Python.

```python
x = [1, 4, 6, 2]
n = len(x)

summation2 = sum(10*x[i] for i in range(0, n))
summation2
```

Explanation

1. `x = [1, 4, 6, 2]`: This line defines a list `x` containing four numbers: 1, 4, 6, and 2.

2. `n = len(x)`: This line calculates the length of the list `x`, which is the number of elements in the list. In this case, `n` will be equal to 4 because there are four elements in the list `x`.

3. `summation2 = sum(10*x[i] for i in range(0, n))`: This line calculates a summation using a loop. Let's break it down further:

   - `range(0, n)` generates a sequence of numbers from 0 to `n-1`. In this case, it generates the sequence [0, 1, 2, 3] because `n` is 4.

   - `x[i]` represents an element of the list `x` at position `i`. For example, `x[0]` is 1, `x[1]` is 4, `x[2]` is 6, and `x[3]` is 2.

   - `10 * x[i]` multiplies each element of `x` by 10. So, it becomes [10, 40, 60, 20].

   - `sum(10 * x[i] for i in range(0, n))` calculates the sum of all the elements in the list obtained in the previous step. It adds up 10 + 40 + 60 + 20, which equals 130.

So, in mathematical terms, this code calculates the sum of each element in the list `x` after multiplying it by 10. The result, `summation2`, is the sum of [10, 40, 60, 20], which is equal to 130.
